# violin plot

```python
n = stats.norm
data = n.rvs(size=(100))

z = stats.norm(loc=3, scale = 1.5)
data1 = z.rvs(size=(100))

df = pd.DataFrame({'Girls': data, 'Boys': data1})

sns.violinplot(df)
```

```python
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import get_test_data

fig = plt.figure(figsize=plt.figaspect(0.5))

ax = fig.add_subplot(1, 2, 1, projection='3d')

# Generate the grid

X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)

# Generate the surface data

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
cmap=cm.GnBu, linewidth=0, antialiased=False)

ax.set_zlim3d(-1.01, 1.01)

fig.colorbar(surf, shrink=0.5, aspect=10)

ax = fig.add_subplot(1, 2, 2, projection='3d')
X, Y, Z = get_test_data(0.05)

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

outfile = '3dGraph.png'
plt.savefig(outfile, dpi=300)
print('Image saved to {0}'.format(outfile))
plt.show()
```

# Bernoulli Distribution

```python
from scipy import stats
p=0.5

bd = stats.bernoulli(p)
p_tails = bd.pmf(0)
p_heads = bd.pmf(1)

trails = bd.rvs(10) #rvs is random variates
trails
```

# Binomial distribution

```python
# if we flip a coin multiple times and ask how often did heads come up? then we have binomial distribution.

from scipy import stats
import numpy as np

(p, num) = (0.5, 4) # for four coin tosses

bd = stats.binom(num,p)
bd.pmf(np.arange(5))
```

# Poisson Distribution

```python
from scipy.stats import poisson
import seaborn as sb

binom = poisson.rvs(mu=4, size=10000)
ax = sb.distplot(binom,
                  kde=True,
                  color='blue',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Poisson', ylabel='Frequency')
```

# Normal Distribution

```python
mu = -2
sigma = 0.5
dist = stats.norm(mu, sigma)
alpha = 0.05

dist.ppf([alpha/2, 1-alpha/2])
```
